```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("./DATASET/canada_per_capita_income.csv")
data.head()
```

```python
data.isnull().sum()
```

```python
plt.scatter(data['year'],data['per capita income (US$)'])
```

```python
from sklearn.linear_model import LinearRegression
```

```python
data
```

```python
lr = LinearRegression()
lr.fit(data[['year']],data['per capita income (US$)'])
```

```python
lr.predict([[2020]])
```

```python
lr.score(data[['year']],data['per capita income (US$)'])
```

```python
plt.scatter(data['year'],data['per capita income (US$)'])
plt.plot(data['year'],lr.predict(data[['year']]))
```

```python
#  STORING MODEL IN THE FILE TO USE IT FURTHER
```

```python
import pickle
with open("model_price","wb") as f:
    pickle.dump(lr,f)
```

```python
with open("model_price","rb") as f:
    model = pickle.load(f)
```

```python
model.predict([[2025]])
```
