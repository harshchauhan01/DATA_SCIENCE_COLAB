```python
data = {'Days':['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'],
        'Week1':[27,29,32,33,33,31,29],
        'Week2':[29,30,31,33,30,31,31],
        'Week3':[29,28,25,27,31,30,29]}
df = pd.DataFrame(data)
df
```

```python
week1_m = np.mean(data['Week1'])
week2_m = np.mean(data['Week2'])
week3_m = np.mean(data['Week3'])
print(week1_m,week2_m,week3_m)
plt.pie([week1_m,week2_m,week3_m],labels=['Week1','Week2','Week3'])
plt.title("Respresentation of Average temperature in particular weeks")
plt.show()
```

```python
# Q4 : Create 5 subplots of 2*2 grid order of trignometric function [Sin, Cos,Tan,Cot,Cosec]
x = np.linspace(-5,5,400)

sin_x =np.sin(x)
cos_x = np.cos(x)
tan_x = np.tan(x)
cot_x = 1/tan_x
cosec_x = 1/sin_x

fig,axes = plt.subplots(2,3,figsize=(15,10))
fig.suptitle("Trigonometric Function")

axes[0,0].plot(x,sin_x,label='Sin(X)')
axes[0,0].set_title("Sin Function")
axes[0,0].legend()

axes[0,1].plot(x,cos_x,label='Cos(X)')
axes[0,1].set_title("Cos Function")
axes[0,1].legend()


axes[0,2].plot(x,tan_x,label='Tan(X)')
axes[0,2].set_title("Tan function")
axes[0,2].legend()

axes[1,0].plot(x,cot_x,label='Cot(X)')
axes[1,0].set_title("Cot Function")
axes[1,0].legend()

axes[1,1].plot(x,cosec_x,label='Cosec(X)')
axes[1,1].set_title("Cosec Function")
axes[1,1].legend()
```
